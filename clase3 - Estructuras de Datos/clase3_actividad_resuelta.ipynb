Ahora que ya tiene manejo sobre las lista vinculadas queremos ver hasta donde puede llegar.

Para esta actividad le pediremos que cree una lista doblemente vinculada, esta es una major a las listas que ya conocemos, pero esta tiene 2 agregados.
La clase Vlist no solo va a contar con un head sino que también con un bottom lo que nos permitira leer la lista de forma invertidad.

Para poder hacer esto en node además de val y next agregararemos un tercer atributo quue es prev, el cual guardar la dirección del nodo que esta antes.

Para la actividad usted tendra que:
- Crear la clase DVList
- Crear la clase DBNode
- La función add_to_front
- La función add_to_back
- La función remove_from_front (elimina el primer valor)
- La función remove_from_back (elimina el ultimo valor)
- La función remove_val (eliminar el primer nodo con el valor dado)
- Print lista
- Print lista de forma inversa




```
class DVList:
  def __init__(self):
    self.head = None
    self.bottom = None

  def __str__(self):
        return 'Lista Vinculada Doble'

  def print_values(self):
    actual = self.head # Variable que apunta al primer nodo de la lista
    while actual != None: # Vamos a revisar la lista hasta que el nodo en el que estamos no tenga un siguiente nodo a revisar
      print(actual.value,end=" ")
      actual = actual.next # Le decimo a actual que tome la información del siguiente nodo

    return

  def print_values_reverse(self):
    actual = self.bottom # Variable que apunta al primer nodo de la lista
    while actual != None: # Vamos a revisar la lista hasta que el nodo en el que estamos no tenga un siguiente nodo a revisar
      print(actual.value,end=" ")
      actual = actual.prev # Le decimo a actual que tome la información del siguiente nodo

    return
  
  def add_to_front(self, val):
    new_node = DBNode(val)
    actual_head = self.head

    if self.head == None:
      self.head = new_node
      self.bottom = new_node
    else:
      self.head = new_node
      new_node.next = actual_head
      actual_head.prev = new_node
    
    return self
  
  def add_to_back(self,val):
    if self.head == None:
      self.add_to_front(val)
      return self
    new_node = DBNode(val)
    actual_bottom = self.bottom
    actual_bottom.next = new_node
    new_node.prev = actual_bottom
    self.bottom = new_node
    return self

  def remove_from_front(self):
    if self.head == None:
      return self
    actual_head = self.head
    new_head = self.head.next
    actual_head = None
    if new_head != None:
      new_head.prev = None
    else:
      self.bottom = None
    self.head = new_head
    return self
  
  def remove_from_back(self):
    if self.head == None:
      return self
    
    actual_bottom = self.bottom
    new_bottom = actual_bottom.prev
    actual_bottom = None
    new_bottom.next = None
    self.bottom = new_bottom
    return self

  def remove_val(self,val):
    if self.head == None:
      return self

    if self.head.value == val:
      self.remove_from_front()
      return self

    actual = self.head
    while actual.next != None:
      if actual.next.value == val:
        break
      if actual.next == None:
        return self

      actual = actual.next

    if actual.next == None:
      return self

    next = actual.next.next
    actual.next = next
    next.prev = actual


    return self


class DBNode:
  def __init__(self,val):
    self.value = val
    self.next = None
    self.prev = None
  
  def __str__(self):
        return f"Node: {self.value}"
my_list = DVList()
my_list.add_to_back(10)
my_list.add_to_front(2)
my_list.add_to_front(3)
my_list.add_to_front(4)
my_list.add_to_back(12)
my_list.add_to_back(13)
my_list.print_values()
print()
my_list.print_values_reverse()
print('\n---')
my_list.remove_from_back()
my_list.print_values()
print()
my_list.print_values_reverse()
print('\n---')
my_list.remove_from_front()
my_list.print_values()
print()
my_list.print_values_reverse()
print('\n---')
my_list.remove_val(10)
my_list.print_values()
print()
my_list.print_values_reverse()
print('\n---')
my_list.remove_val(3)
my_list.print_values()
print()
my_list.print_values_reverse()
print('\n---')
my_list.remove_val(2)
my_list.print_values()
print()
my_list.print_values_reverse()
print('\n---')
my_list.remove_val(12)
my_list.print_values()
print()
my_list.print_values_reverse()
```
